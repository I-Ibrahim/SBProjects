Define function to correct spelling based on enchant package, and capitalises letters in sentence

```python
d=enchant.Dict("en_UK")

def correct_spelling(x):
    sentence = re.split('([.!?] *)', x)
    final = ''.join([i.capitalize() for i in sentence])
    
    final=final.split()
    corrected=[]
    for i in final:
        correct=d.check(i)
        if correct==False:
            i=d.suggest(i)[0]
            corrected.append(i)
        else:
            corrected.append(i)
    corrected=" ".join(corrected)
    return corrected
```

Test function on rows selected above (row 2 and 521 in same column)

```python
print (correct_spelling(df['questions_48542_why_is_this_your_preferred_brand'][2]))
print (correct_spelling(df['questions_48542_why_is_this_your_preferred_brand'][521]))
```
